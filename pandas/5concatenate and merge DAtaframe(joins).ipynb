```python
import pandas as pd
```

```python
# 1. Concatenate (DataFrames ko jodna)

# 👉 concat() ka use hota hai:

# Multiple DataFrames ko stack (upar-neeche ya side-by-side) jodne ke liye
# pd.concat([df1, df2])
```

```python
df1 = pd.DataFrame({
    'name':['ram','ravi'],
    'age':[25,30]
})
df2= pd.DataFrame({
    'name':['amit','sita'],
    'age':[28,22]
})
#result=pd.concat([df1,df2],axis =0) # c
result=pd.concat([df1,df2],axis=0) # concatenate:horizontal/column/level sql joins
```

```python
df1 = pd.DataFrame({
    'empid': [1,2,3],
    'name': ['ram','ravi','amit']
})
df2 =pd.DataFrame({
    'empid':[1,2,4],
    'salary':[30000,40000,50000]
})
    
```

```python
pd.merge(df1,df2,on='empid') #inner join (defaoult ) sirf matching data
```

```python
#left join tablle ka sab data + matching right
pd.merge(df1,df2,on='empid',how='left') # left join
```

```python
#right join table ka sab data + matching left

pd.merge(df1,df2,on='empid',how='right') # right join
```

```python
# outer join sab data show karega table ka 
pd.merge(df1,df2,on ='empid',how='outer')
```

```python

# Join Type	Result
# Inner	Only matching
# Left	Left + matching right
# Right	Right + matching
# Outer	All data
```

```python
# Dono DataFrame ko empid column ke basis par inner join se merge kiya


pd.merge(df1,df2,how='inner',left_on='empid',right_on='empid')
```
