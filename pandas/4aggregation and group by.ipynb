```python
# 🔷 1. GroupBy kya hota hai?

# 👉 groupby() ka matlab:

# “Same category wale data ko group karo”

# Example:

# Department wise salary
# City wise sales
# Gender wise marks

# df.groupby('column_name')
```

```python
import pandas as pd
```

```python
data = {
    'Name': ['Ram', 'Ravi', 'Amit', 'Ram', 'Ravi'],
    'Dept': ['IT', 'HR', 'IT', 'HR', 'IT'],
    'Salary': [30000, 25000, 40000, 32000, 28000]
}

df = pd.DataFrame(data)
```

```python
print(df)
```

```python
# 2. Aggregation kya hota hai?
# 👉 Aggregation ka matlab:

# “Grouped data par calculation karna”

# Common functions:

# sum()
# mean()
# min()
# max()
# count()
```

```python
df['Dept'].value_counts() # frequency values in month column
```

```python
df[df['Dept']=='IT'].value_counts() #frequncy 
```

```python
#depatrtment wise total salary[]
df.groupby('Dept')['Salary'].sum()
```

```python
#department wise total avg salary
df.groupby('Dept')['Salary'].mean()
```

```python
#count of employees per department
df.groupby('Dept')['Name'].count()
```

```python
#multiple aggregations(most important)

df.groupby('Dept')['Salary'].agg(['sum','mean','min','max'])
```

```python
#multiple column group by  dept +name combination wise result
df.groupby(['Dept','Name'])['Salary'].sum()
```

```python
#name aggregation (advanced)
df.groupby('Dept').agg(
    total_salary=('Salary','sum'),
    avg_salary =('Salary','mean'),
    count_emp=('Name','count') )
```

```python
df.to_csv('data2.csv')
```

```python
# ⚡ 8. Simple Trick to Remember

# 👉 GroupBy = “Category banana”
# 👉 Aggregation = “Calculation karna”


# Concept	Meaning
# groupby()	Data ko groups me divide karta hai
# aggregation	grouped data par calculation karta hai
# agg()	multiple functions ek saath

# Pandas me GroupBy + Aggregation = Data Analysis ka backbone hai
```

```python
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Laptop', 'Mouse', 'Keyboard', 'Laptop'],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South', 'North'],
    'Sales': [50000, 500, 1500, 55000, 600, 1400, 60000],
    'Profit': [8000, 100, 300, 9000, 120, 280, 10000]
}

df = pd.DataFrame(data)
df
```

```python
#product wise sales
df.groupby('Product')['Sales'].sum()
```

```python
#product wise average profit
df.groupby('Product')['Profit'].mean()
```

```python
#region wise tatal sales
df.groupby('Region')['Sales'].sum()
```

```python
# region wise profit
df.groupby('Region')['Profit'].sum()
```

```python
#best selling product
df.groupby('Product')['Sales'].sum().sort_values(ascending=False)
```

```python
#full bussiness report (most imp)
report = df.groupby('Product').agg(
    total_sales=('Sales', 'sum'),
    avg_sales=('Sales', 'mean'),
    total_profit=('Profit', 'sum'),
    avg_profit=('Profit', 'mean')
)

report
```

```python
#top performaing region
df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Product')['Sales'].sum().sort_values(ascending=True)
```
