```python
import pandas as pd
```

```python
data = {
    'name':['ram','ravi','ak','vj','av','rt'],
    'age':[26,37,28,19,12,23],
    'salary':[1000,3000,4000,5000,2000,3000]
}

df = pd.DataFrame(data)

print(df)
print(type(df))
```
