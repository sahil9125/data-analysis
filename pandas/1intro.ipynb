```python
s = pd.Series([10,20,30,40,50])
print(s)
#print(s[0])
```

```python
# custom index
s = pd.Series([100,200,300], index =['a', 'b' ,'c'])
print(s)
print(s['b'])
```

```python
#dictionary se series
```

```python
data ={
    'name':'ravi',
    'course':'python',
    'fees':5000
}
s = pd.Series(data)
print(s)
print(s.values)
print(s.index)
print(s.dtype)
```

```python
# Difference: List vs Series
# List	Series
# Normal Python   	Pandas Object
# No label	        Index hota hai
# Less powerful	    Data analysis ke liye useful

# Series vs DataFrame
# Series	      DataFrame
# Single Column	  Multiple Columns
# 1D	          2D
```

```python
# What is Dataframe?
# DataFrame Pandas ka ek 2-dimensional data structure hai jo data ko rows aur columns ke form mein 
#store karta hai,# bilkul Excel sheet ya database table ki tarah.

# DataFrame ek table hota hai jisme multiple rows aur columns hote hain.

# | Name   | Age | City    |
# | ------ | --- | ------- |
# | Aman   | 22  | Gwalior |
# | Rahul  | 24  | Indore  |
# | Simran | 21  | Bhopal  |
```

```python
df= pd.DataFrame([10,11,12],columns=['Col_Nmae']) # square Brackets() parentheses
print(df) # 0 kiyaki kayi col ka name nahi diya  hai is liye 0 de raha hai
print(type(df))
```
