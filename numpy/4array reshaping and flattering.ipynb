```python
import numpy as np
```

```python
#reshape array:array ke data ko same rakhte hue ushka structure change karna 
#r latter arey : 2d ya 3dko ek single(1d) me convert karna.arr
```

```python
arr1 = np.array([1,2,3,4,5,6])
print(arr1)   # shape 1row and 6 column
```

```python
reshaped = arr1.reshape(6,1)  #6 row and 1 column
print(reshaped)
```

```python
reshaped2 = arr1.reshape(2,3)  # re shaped array from (1,6 )to (2,3)
print(reshaped2)
```

```python
#flatten array : 2d ya 3d ko ek single line (1d) convert karna.
print (reshaped2.flatten()) # fallened array from 2d dimensional (2,3) to (1,6) (1d)
```

```python
#array stacking : do ya zyada array ko jodna (combine karna ). 
#type 2
#1)(vertical.stack np.vstack(a,b)))
#2)horizontal stack: np.hstack(a,b))
#array splitting: ek array ko chhote part mai todna
```

```python
#stacking
a = np.array([1,2,3])
b = np.array([4,5,6])
stacked = np.stack((a,b))
print(stacked)

print(np.vstack((a,b)))  # vertical stacking - row wise output 2d
print(np.hstack((a,b))) # horizonatl stack - column wise  output 1d 
```

```python
#splitting
c = np.array([[1,2,3],[4,5,6]])
print(c)
```

```python
hsplit = np.hsplit(c,3) #horizontal split ->input and output array are same dimention
print(hsplit)
for s in hsplit:
    print(s)
```
