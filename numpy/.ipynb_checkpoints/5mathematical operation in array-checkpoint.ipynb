```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a)
print(b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

```python
# dot product : computer the inner product of the two vector and sum the result
print (np.dot(a,b)) # dot product 4 10 18 = 32
```

```python
print(a.T) # transpose : reture an array with  axes transposed
```

```python
t = np.array([[1,2,3],
              [4,5,6]])
print(t)   # initial shape :2,3
print(t.T)  # shape after transpose: 32 row ko col ko row
```

```python
# statistical function
a= np.array ([[1,2,3],[4,5,6]])
print(a)
```

```python
print(np.sum(a))  # ssum of all element
```

```python
print(a)
print(np.mean(a))  #average
print(np.median(a)) #middle value
print(np.std(a))  # standard deviation
print(np.min(a))  # smallest
print(np.max(a))  # largest
```

```python
a = np.array([1,5,3])
b = np.array([4,5,6])
print(a)
print(b)
```

```python
print(a==b) # comparing each element of array
print(np.array_equal(a,b)) # comapring complete array
```
