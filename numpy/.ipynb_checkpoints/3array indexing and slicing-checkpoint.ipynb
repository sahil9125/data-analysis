```python
import numpy as np
```

```python
a = np.array([1,3,5,7,9])
print(a)
```

```python
print(a[0]) #first element of array
```

```python
print(a[4])
```

```python
print(a[-1])
```

```python
a = np.array([1,3,5,7,9])
print(a)
#slicing:
#array[start:stop:step]
#stop  - index , that  is not included in output
```

```python
print(a[0:3]) #first 3 elemnts
```

```python
print(a[-3:]) # last 3 elemnts
```

```python
print(a[0::3])  # midle 3 elments with steps
```

```python
print(a[0::2])
```

```python
# indexing on 2 dimensional array 
arr2 = np.array([[1,2,3], [4,5,6]])
print(arr2)
#2d array {row :column}
#2d array takes 2 value - row and column for indexing
```

```python
print(arr2[0])  # first row
print(arr2[1])  # second row
```

```python
print(arr2[0][2])  # first row
print(arr2[1][2] ) # second row
```

```python
print(arr2[1:])
```
