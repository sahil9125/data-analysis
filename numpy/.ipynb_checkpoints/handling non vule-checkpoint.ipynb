```python
import numpy as np
#handling with non value
```

```python
data1 = np.array([1,2,np.nan,4,np.inf])
print(data1)
```

```python
print(np.isnan(data1))
```

```python
print(np.nan_to_num(data1)) #null value ko 0 mai badlta 
```

```python
#save and load arrays
```

```python
arr =np.array([1,0,0,8])
print(arr)
np.save('my_array.npy',arr)
```

```python
loaded_arr=np.load('my_array.npy')
print(loaded_arr)
```

```python
# create 5x5 martrix 
```

```python
arr=np.arange(1,26).reshape(5,5)
print("orginal matrix:\n;",arr)
```

```python
arr = np.arange(1,26).reshape(5,5)
print("original matrix:\n",arr)
print("last row:\n",arr[-1])
print("first column:\n",arr[:,0])
print("3x submatrix:\n",arr[1:4,1:4])
```
