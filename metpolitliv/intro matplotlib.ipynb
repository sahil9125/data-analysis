```python
import matplotlib.pyplot as plt
# pyplot Matplotlib ka ek module hai jo graph banane ke liye simple functions provide karta hai.
# matplotlib = poori library
# pyplot = graph banane wala module
# plt = short name (alias)
```

```python
#data 
x =[1,2,3,4,5]
y =[10,20,25,35,45]
plt.plot(x,y)  #line chart
plt.show() #display chart in jp nb
```

```python
plt.figure(figsize=(4,3)) # figure size inch
plt.plot(x, y, color='y', marker='o', linestyle='--', linewidth=2, markersize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,3))  # figure size

plt.plot(
    x, y,
    color='y',
    marker='o',
    linestyle='--',
    linewidth=2,
    markersize=10
)

plt.title("MY FIRST GRAPH")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
#data
x=[1,2,3,4,5]
y1=[10,20,30,39,55]
y2=[20,35,34,57,78]
plt.plot(x,y1,label='sales 2022',marker='o')
plt.plot(x,y2,label='sales 2023',marker='>')

plt.title('PN SALES')
plt.xlabel('months')
plt.ylabel('sales')
plt.grid()

plt.legend() #show legends
plt.show()
```

```python
months =[1,2,3,4,5]
sales=[10,20,30,33,20]
profit=[23,30,40,50,10]

plt.plot(months ,sales ,label='sales')
plt.plot(months ,profit ,label='profit')

plt.legend()
plt.show
```

```python
#line color ,style &width
plt.plot(
    months,
    sales,
    color='red',
    linestyle='-.',
    linewidth=3
    
)
plt.show()
#'-' solid line
#'--' dashed line
#':' dotted line
#'-.' dash dot
```

```python
#marker use karna
plt.plot(
    months,
    sales,
    marker='^'
)
plt.show()
#'o' circle
#'s' square
#'^' aroow
```

```python
#title & labels
plt.plot(months ,sales)
plt.title('monthly sales')
plt.xlabel('months')
plt.ylabel('sales')

plt.show()
```

```python
#figure size control
plt.figure(figsize=(10,5))

plt.plot(months,sales)
plt.grid()

plt.show()
#width =10
#height =5
```

```python
#data point annotate karna 
plt.plot(months,sales,marker='o')
for x,y in zip(months ,sales):
    plt.text(x,y,str(y))
plt.show()
#har point ke upr vlaue dikhegi.
```

```python
# Pandas DataFrame ke Sath
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Month':[1,2,3,4,5],
    'Sales':[10,20,15,5,10]
})

plt.plot(df['Month'],df['Sales'],marker='*')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid()


plt.show()
# Real Data Analysis me isi tarah use hota hai.
```

```python

# Professional Advanced Example
months = ['Jan','Feb','Mar','Apr','May']

sales = [10000,15000,18000,25000,35000]
profit = [3000,5000,7000,10000,15000]

plt.figure(figsize=(10,5))

plt.plot(months,sales,
         marker='o',
         linewidth=3,
         label='Sales')

plt.plot(months,profit,
         marker='s',
         linewidth=3,
         label='Profit')

plt.title("Sales vs Profit Analysis")
plt.xlabel("Months")
plt.ylabel("Amount")
plt.grid(True)
plt.legend()

plt.show()
```
