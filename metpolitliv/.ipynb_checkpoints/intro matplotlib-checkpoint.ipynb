```python
import matplotlib.pyplot as plt
# pyplot Matplotlib ka ek module hai jo graph banane ke liye simple functions provide karta hai.
# matplotlib = poori library
# pyplot = graph banane wala module
# plt = short name (alias)
```

```python
#data 
x =[1,2,3,4,5]
y =[10,20,25,35,45]
plt.plot(x,y)  #line chart
plt.show() #display chart in jp nb
```

```python
#customize chart
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
```

```python

plt.plot(x, y, color='darkgreen', marker='o', linestyle='--', linewidth=2, markersize=10)
plt.show()
```

```python
plt.figure(figsize=(4,3)) # figure size inch
plt.plot(x, y, color='y', marker='o', linestyle='--', linewidth=2, markersize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,3))  # figure size

plt.plot(
    x, y,
    color='y',
    marker='o',
    linestyle='--',
    linewidth=2,
    markersize=10
)

plt.title("MY FIRST GRAPH")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
#advanced - multiple line &legends line chart
```

```python
#data
x=[1,2,3,4,5]
y1=[10,20,30,39,55]
y2=[20,35,34,57,78]
plt.plot(x,y1,label='sales 2022')
plt.plot(x,y2,label='sales 2023')

plt.title('PN SALES')
plt.xlabel('months')
plt.ylabel('sales')

plt.legend() #show legends
plt.show()
```

```python
months =[1,2,3,4,5]
sales=[10,20,30,33,20]
profit=[23,30,40,50,10]

plt.plot(months ,sales ,label='sales')
plt.plot(months ,profit ,label='profit')

plt.legend()
plt.show
```

```python
#line color ,style &width
plt.plot(
    months,
    sales,
    color='red',
    linestyle='-.',
    linewidth=3
    
)
plt.show()
#'-' solid line
#'--' dashed line
#':' dotted line
#'-.' dash dot
```

```python
#marker use karna
plt.plot(
    months,
    sales,
    marker='^'
)
plt.show()
#'o' circle
#'s' square
#'^' aroow
```

```python
#title & labels
plt.plot(months ,sales)
plt.title('monthly sales')
plt.xlabel('months')
plt.ylabel('sales')

plt.show()
```

```python
#figure size control
plt.figure(figsize=(3,5))

plt.plot(months,sales)

plt.show()
#width =10
#height =5
```
