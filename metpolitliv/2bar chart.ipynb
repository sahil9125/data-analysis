```python
import matplotlib.pyplot as plt
```

```python
students = ["ram",'shyam',"mohan"]
marks =[80,90,70]
plt.bar(students,marks)
plt.show()
```

```python
#title aur labels
plt.bar(students , marks)
plt.title("student marks")
plt.xlabel("students")
plt.ylabel("marks")
plt.show()
```

```python
#bar color
plt.bar(students ,marks,color="y")
plt.show()
```

```python
#bar width
plt.bar(
    students,
    marks,
    width=0.1
)
plt.show()
```

```python
#grid add karna
plt.bar(students,marks)
plt.grid(axis='y') # bar chart me mostly axis =y use hota hai.

plt.show()
```

```python
students = ["Ram", "Shyam", "Mohan"]
marks = [80, 90, 70]

bars = plt.bar(students, marks)

plt.bar_label(bars,padding=2)   # Value automatically bar ke upar

plt.show()
```

```python
# Horizontal Bar Chart
plt.barh(students, marks) #Horizontal Bar Chart

plt.show()
```

```python
colors = ['red','green','blue']

plt.bar(
    students,
    marks,
    color=colors
)

plt.show()
```

```python
# Figure Size
plt.figure(figsize=(8,5))

plt.bar(students, marks)

plt.show()
```

```python
# Grouped Bar Chart
# Grouped Bar Chart ka use ek hi category ke andar multiple data sets ko compare karne ke liye kiya jata hai, 
# jaise Students ke Maths aur Science marks.


import numpy as np
students =["ram","shyam","mohan"]
maths=[80,90,70]
science=[75,85,65]
x =np.arange(len(students))

plt.bar(x-0.2,maths,width=0.4,label="maths")
plt.bar(x+0.2,science,width=0.4,label="science")


plt.xticks(x,students)

plt.legend()

plt.show()


# Ek student ke liye 2 subjects compare kar sakte ho.
# x = [0,1,2] → categories ki positions
# x-0.2 → left shift
# x+0.2 → right shift
# width=0.4 → bar ki width
# plt.xticks() → numbers ko names me badalta hai
# plt.legend() → Maths aur Science ka label dikhata ha
```

```python
# Stacked Bar Chart
students = ["Ram","Shyam","Mohan"]

maths = [80,90,70]
science = [20,10,15]

plt.bar(students, maths, label='Maths')

plt.bar(
    students,
    science,
    bottom=maths,
    label='Science'
)

plt.legend()

plt.show()
# Yahan bars ek doosre ke upar stack hoti hain.
```

```python
# Pandas ke Sath
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Student':['Ram','Shyam','Mohan'],
    'Marks':[80,90,70]
})

df.plot(
    kind='bar',
    x='Student',
    y='Marks'
)

plt.show()
```
