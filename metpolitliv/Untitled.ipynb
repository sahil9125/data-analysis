```python
# # Subplots Kya Hote Hain?

# # Subplots ka matlab hai:

# # Ek hi figure (window) me multiple graphs dikhana.
# # plt.plot(...)
# # plt.show()
# # Ek graph dikhata hai.
# # Real Life Example

# # Maan lo aap monthly report bana rahe ho:

# # Sales Chart
# # Profit Chart
# # Expense Chart
# # Customer Chart

# # Sab alag-alag window me dikhane ki jagah:
# # +----------+----------+
# # | Sales    | Profit   |
# # +----------+----------+
# # | Expense  | Customer |
# # +----------+----------+
# # Ek hi figure me dikhate hain.
# # plt.subplot(rows, columns, position)
# याद रखो:

# Row = Horizontal line (↔)
# Column = Vertical division (↕)

# 1 row, 2 column = एक लाइन में 2 graph की जगह।
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3],[4,5,6])

plt.subplot(1,2,2)
plt.plot([1,2,3],[3,2,1])

plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot([1,2,3],[1,2,3])

plt.subplot(2,2,2)
plt.plot([1,2,3],[3,2,1])

plt.subplot(2,2,3)
plt.plot([1,2,3],[2,3,1])

plt.subplot(2,2,4)
plt.plot([1,2,3],[1,3,2])

plt.show()
```

```python
# Graph 1
plt.subplot(2,2,1)
plt.plot([1,2,3],[1,2,3])
plt.title("Sales")
plt.xlabel("Month")
plt.ylabel("Amount")

# Graph 2
plt.subplot(2,2,2)
plt.plot([1,2,3],[3,2,1])
plt.title("Profit")
plt.xlabel("Month")
plt.ylabel("Amount")

# Graph 3
plt.subplot(2,2,3)
plt.plot([1,2,3],[2,3,1])
plt.title("Expenses")
plt.xlabel("Month")
plt.ylabel("Amount")

# Graph 4
plt.subplot(2,2,4)
plt.plot([1,2,3],[1,3,2])
plt.title("Customers")
plt.xlabel("Month")
plt.ylabel("Count")

plt.tight_layout()

plt.show()
```

```python
months = ['Jan','Feb','Mar','Apr']

sales = [100,150,200,250]
profit = [20,40,60,80]

plt.subplot(1,2,1)
plt.plot(months,sales)
plt.title("Sales")

plt.subplot(1,2,2)
plt.plot(months,profit)
plt.title("Profit")

plt.show()
```

```python
# # Meaning
# # fig

# # Pure figure ko represent karta hai.

# # ax

# # Graphs ko represent karta hai.
# Step 1: Import
# import matplotlib.pyplot as plt

# Matplotlib library import की।

# Step 2: Create Subplots
# fig, ax = plt.subplots(2,2)

# इस लाइन में 2 चीजें बनती हैं:

# fig
# fig

# Figure = पूरी graph window

# इसे पूरा canvas समझो।

# +-----------------------+
# |                       |
# |      Figure           |
# |                       |
# +-----------------------+
# ax
# ax

# Axes = Graph area

# क्योंकि हमने:

# plt.subplots(2,2)

# लिखा है।

# इसलिए:

# 2 Rows
# 2 Columns

# यानि 4 graph areas बनेंगे।

# +-----------+-----------+
# | ax[0,0]   | ax[0,1]   |
# +-----------+-----------+
# | ax[1,0]   | ax[1,1]   |
# +-----------+-----------+
# 2,2 का मतलब
# plt.subplots(2,2)

# मतलब:

# 2 Rows
# 2 Columns

# Layout:

# +---------+---------+
# |    1    |    2    |
# +---------+---------+
# |    3    |    4    |
# +---------+---------+

# Total:

# 2 × 2 = 4 Graphs
```

```python
fig, ax=plt.subplots(2,2)
ax[0,0].plot([1,2,3],[1,2,3])

plt.show()
#ax[0,0]
#ax[0,0]
#row 0
#column 0
```

```python
fig, ax=plt.subplots(2,2)
ax[0,0].plot([1,2,3],[1,2,3])
ax[0,1].plot([1,2,3],[3,2,1])

ax[1,0].plot([1,2,3],[3,2,1])
ax[1,1].plot([1,2,3],[1,3,2])


plt.show()
```

```python
fig, ax = plt.subplots(2,2)

ax[0,0].plot([1,2,3],[1,2,3])
ax[0,1].plot([1,2,3],[3,2,1])

ax[1,0].plot([1,2,3],[2,3,1])
ax[1,1].plot([1,2,3],[1,3,2])

plt.show()
```

```python
onths = ['Jan','Feb','Mar','Apr']

sales = [100,150,200,250]
profit = [20,40,60,80]

fig, ax = plt.subplots(1,2, figsize=(10,5))

ax[0].plot(months,sales)
ax[0].set_title("Sales")

ax[1].plot(months,profit)
ax[1].set_title("Profit")

plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(10,8))

# 1. Line Chart
ax[0,0].plot([1,2,3,4],[10,20,15,25])
ax[0,0].set_title("Line Chart")

# 2. Bar Chart
ax[0,1].bar(
    ["Jan","Feb","Mar","Apr"],
    [100,150,120,180]
)
ax[0,1].set_title("Bar Chart")

# 3. Pie Chart
ax[1,0].pie(
    [40,30,20,10],
    labels=["Python","Java","MERN","DA"],
    autopct='%1.1f%%'
)
ax[1,0].set_title("Pie Chart")

# 4. Scatter Plot
ax[1,1].scatter(
    [1,2,3,4,5],
    [20,40,35,60,80]
)
ax[1,1].set_title("Scatter Plot")

plt.tight_layout()
plt.show()
```
