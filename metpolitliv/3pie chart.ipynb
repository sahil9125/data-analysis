```python
import matplotlib.pyplot as plt

# # Pie Chart Kya Hota Hai?

# # Pie Chart ek circular chart hota hai jo data ko percentage ya parts me dikhata hai.

# # Iska use tab karte hain jab hume dekhna ho:

# # Total me kis category ka kitna contribution hai.
# Example

# Ek class me:

# Python = 40 students
# Java = 30 students
# MERN = 20 students
# Data Analytics = 10 students

# Total:

# 40 + 30 + 20 + 10 = 100

# Pie Chart dikhayega:

# Python = 40%
# Java = 30%
# MERN = 20%
# Data Analytics = 10%
```

```python
courses =['python' , 'java','mern','data analytics']
students =[40,30,20,10]
plt.pie(students)
plt.show()
```

```python
#label add karna
# agr  labels nahi dege to srif slices dikhenge .har slice ke sath naam dikhega.
plt.pie(
    students,
    labels=courses
)
plt.show()
```

```python
plt.pie(
    students,
    labels=courses,
    autopct='%1.1f%%'
)
plt.title("Course Enrollment Distribution")

plt.show()
```

```python
courses = ["Python", "Java", "MERN", "Data Analytics"]

students = [40, 30, 20, 10]

plt.pie(
    students,
    labels=courses,
    autopct='%1.1f%%'
)

plt.title("Course Enrollment Distribution")

plt.show()
```

```python
explode=[0,0.1,0,0] # 0.1 0.1
plt.pie(
    students,
    labels=courses,
    autopct='%1.1f%%',
    explode=explode
    
)
plt.show()
```

```python
colors = ['yellow', 'blue', 'green', 'orange']

plt.pie(
    students,
    labels=courses,
    colors=colors,
    autopct='%1.1f%%'
)

plt.show()
```

```python
plt.pie(
    students,
    labels=courses,
    autopct='%1.1f%%',
    shadow=True
)

plt.show()
```

```python
plt.pie(
    students,
    labels=courses,
    autopct='%1.1f%%',
    startangle=90
)

plt.show()
```

```python
# autopct='%1.1f%%'

# Pie Chart में percentage show करने के लिए use होता है।

# Example
# values = [40, 30, 20, 10]

# Total:

# 40 + 30 + 20 + 10 = 100

# तो Pie Chart में:

# 40%
# 30%
# 20%
# 10%

# दिखेगा।

# autopct का मतलब
# auto = automatic
# pct  = percentage

# मतलब:

# Percentage automatically calculate karke show karo.

# '%1.1f%%' को तोड़कर समझो
# '%1.1f%%'
# 1. f
# f

# Floating point number (decimal value)

# Example:

# 40.0
# 30.0
# 20.0
# 2. .1
# .1

# Decimal ke baad 1 digit

# Example:

# 40.0
# 30.0
# 20.0
# 3. %
# %%

# Actual % symbol print karne ke liye.

# Example:

# 40.0%
# 30.0%
# 20.0%
# Example
# import matplotlib.pyplot as plt

# labels = ['Python', 'Java', 'MERN']
# values = [50, 30, 20]

# plt.pie(
#     values,
#     labels=labels,
#     autopct='%1.1f%%'
# )

# plt.show()

# Output:

# Python 50.0%
# Java   30.0%
# MERN   20.0%
# Agar .0 likho
# autopct='%1.0f%%'

# Output:

# 50%
# 30%
# 20%

# Decimal nahi aayega.

# Agar .2 likho
# autopct='%1.2f%%'

# Output:

# 50.00%
# 30.00%
# 20.00%

# 2 decimal places.

# Yaad Rakhne Ka Shortcut
# Format	Output
# %1.0f%%	40%
# %1.1f%%	40.0%
# %1.2f%%	40.00%
# %1.3f%%	40.000%

# Most commonly used:

# autopct='%1.1f%%'
```

```python
# # Pie Chart Parameters
# # | Parameter  | Use                 |
# # | ---------- | ------------------- |
# # | labels     | Slice names         |
# # | autopct    | Percentage show     |
# # | explode    | Slice bahar nikalna |
# # | colors     | Slice color         |
# # | shadow     | Shadow effect       |
# # | startangle | Starting angle      |

# Pie Chart vs Bar Chart
# Bar Chart
# students = [40,30,20,10]

# Category comparison.

# Python ██████████
# Java   ████████
# MERN   █████
# DA     ███
# Pie Chart
# Total = 100%

# Shows share of each category in the total.

# Python = 40%
# Java = 30%
# MERN = 20%
# DA = 10%
```
